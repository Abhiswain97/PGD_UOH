# Exploratory Data Analysis for Haberman

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./haberman.csv", names=["Age", "Year", "Positive nodes", "Survival"]) 
```

```python
df.head()
```

## Attribute Information:

1. Age of patient at time of operation (numerical)
2. Patient's year of operation (year - 1900, numerical)
3. Number of positive axillary nodes detected (numerical)
4. Survival status (class attribute)
    - 1 = the patient survived 5 years or longer
    - 2 = the patient died within 5 year

```python
df.shape
```

We have 306 samples, with 3 features

## Scatter plot 

```python
sns.set_style("whitegrid");
_ = sns.FacetGrid(df, hue="Survival", height=4).map(plt.scatter, "Age", "Positive nodes").add_legend()
```

There's a lot of overlap, it seems to be quite uniform for all ages. 

We can conclude that, at whatever age they have breast cancer, the number of nodes stays uniform

next, let's count the number of people who did and didn't survive 5 years

```python
df['Survival'].value_counts()
```

The dataset is highly imbalanced

```python
sns.set_style("whitegrid");
_ = sns.FacetGrid(df, hue="Survival", height=4).map(plt.scatter, "Age", "Year").add_legend()
```

### Observations

- This is also amazingly uniform. 
- We can see that, across `Year` axis for every year there are a considerable number of people 
across all *ages* operating for breast cancer
- Using features `Age` & `Year` or `Age` & `Positive nodes` we can hardly make a separation for survival 

## Pairplot

```python
sns.set_style("whitegrid");
sns.pairplot(df, 
             hue="Survival", 
             height=4, 
             palette=["green", "red"]);
plt.show()
```

### Observations:

- From the pairplot, we can clearly see there's a huge amount of overlap between all the features. 
- No one feature seems to be good enough for us to use for making a decision on the survival 

## Histogram, PDF, CDF

```python
sns.FacetGrid(df, hue="Survival", height=5).map(sns.histplot, "Age").add_legend();
plt.show()
```
